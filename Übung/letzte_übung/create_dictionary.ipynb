```python
datei = open("coords.txt")
Lx=list()
Ly=list()
Dic={}
for zeile in datei:
    if zeile =="X, Y\n":
        x, y= zeile.split(",")
        continue
    Lx.append(float(zeile.split(",")[0]))
    Ly.append(float(zeile.split(",")[1].split("\n")[0]))
Dic[x]=Lx
Dic[y]=Ly
    
Dic
```
